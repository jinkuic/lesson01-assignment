```python
url = 'http://www.bjsubway.com/station/zjgls/#'
```

```python
response = requests.get(url)
response.encoding = 'gbk'
print(response.text)
```

```python
what_we_want = r'<th>(.*?)</th>'
#what_we_want = r'<p><a class="cl-station"'
#what_we_want = r'<a href="https://(movie\.douban\.com/subject/\d+/\?from=subject-page)" class="" '
```

```python
pattern = re.compile(what_we_want)
```

```python
likes = pattern.findall(response.text)
```

```python
station_connection = defaultdict(list)
```

```python
for like in likes:
    pattern = re.compile(r"(\w+)——(\w+)")
    station_info = pattern.findall(like)
    if not station_info: continue
     
    s1, s2 = station_info[0]
    station_connection[s1].append(s2)
    station_connection[s2].append(s1)
```

```python
print(station_connection)
```

```python
def transfer_stations_first(pathes): 
    return sorted(pathes, key=len)
```

```python
def shortest_path_first(pathes):
    
    if len(pathes) <= 1: return pathes
    
    def get_path_distnace(path):
        distance = 0
        for station in path[:-1]:
            distance += 1
            
        return distance

    return sorted(pathes, key=get_path_distnace)
```

```python
def search(start, destination, connection_grpah, sort_candidate):
    pathes = [[start]]
    
    visitied = set()
    
    while pathes: # if we find existing pathes
        path = pathes.pop(0)
        froninter = path[-1]
        
        if froninter in visitied: continue
            
        successors = connection_grpah[froninter]
        
        for city in successors:
            if city in path: continue  # eliminate loop
                
            new_path = path + [city]
            
            pathes.append(new_path)
            
            if city == destination: return new_path
        
        visitied.add(froninter)
        
        pathes = sort_candidate(pathes) # 我们可以加一个排序函数 对我们的搜索策略进行控制
```

```python
search('八宝山', '车公庄', station_connection, sort_candidate=shortest_path_first)
```
